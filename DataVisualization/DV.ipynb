```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple plot')
plt.show()
```

```python
#creting multiple plots
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z = [1, 8, 27, 64, 125]
plt.Figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.plot(x, y , color='red')
plt.title('Plot 1')
plt.subplot(2, 2, 2)
plt.plot(x, z, color='green')
plt.title('Plot 2')
plt.subplot(2, 2, 3)
plt.plot(y, x, color='blue')
plt.title('Plot 3')
plt.show()
```

```python
#Bar chart
categories = ['apple', 'banana', 'grapes', 'mango', 'orange']
values = [23, 45, 56, 78, 33]
plt.bar(categories, values , color='blue')
```

```python
#histogram
data = [1, 2, 1, 3, 3, 1, 4, 2 ,5,5,3,4,1]
plt.hist(data, bins=5, color='orange' , edgecolor='black')
```

```python
#scatter plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.scatter(x, y, color='red' , marker='x')
```

```python
#pie chart
sizes = [20, 30, 25, 25]
labels = ['apple', 'banana', 'mango', 'orange']
colors = ['red', 'green', 'blue', 'yellow']
explode = (0.1, 0, 0, 0)
plt.pie(sizes, labels=labels, colors=colors, explode=explode, shadow=True , autopct='%1.1f%%')
```

```python
import pandas as pd
data = pd.read_csv('data.csv')
data.head(5)
```

```python
total_sales = data.groupby('Product')['Sales'].sum()
print(total_sales)
```

```python
total_sales.plot(kind='bar')
```

```python
sales_trent = data.groupby('Date')['Sales'].sum().reset_index()
```

```python
print(sales_trent)
```

```python
plt.plot(sales_trent['Date'], sales_trent['Sales'])
```
